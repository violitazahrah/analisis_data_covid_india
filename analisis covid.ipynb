```python
covid19_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid19_df.head()
```

```python
covid19_df['Date'] = pd.to_datetime(covid19_df['Date'], format = '%Y-%m-%d')
```

```python
covid19_df.head()
```

```python
# kasus aktif

covid19_df['Active_cases'] = covid19_df['Confirmed'] - (covid19_df['Cured'] + covid19_df['Deaths'])
```

```python

import random
middle_rows = covid19_df.sample() 
covid19_df.sample(5)
```

```python
statewise = pd.pivot_table(covid19_df, values = ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = max)
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# Top 10 Kasus active

top_10_active_cases = covid19_df.groupby(by ='State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by = ['Active_cases'], ascending = False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 wilayah Covid Tertinggi di India", size = 25 )
```

```python
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'red')
```

```python
# Top 10 Kasus aktif
top_10_active_cases = covid19_df.groupby(by ='State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by = ['Active_cases'], ascending = False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 wilayah Covid Tertinggi di India", size = 25 )
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'red')

plt.xlabel("Wilayah")
plt.ylabel("Total kasus aktif")
plt.show()
```

```python
#Top 10 wilayah jumlah kematian tertinggi
top_10_deaths = covid19_df.groupby(by ='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()

fig = plt.figure(figsize=(18,5))

plt.title("Top 10 Wilayah Kematian Covid Tertinggi di India", size =25)

ax = sns.barplot(data = top_10_deaths.iloc[:12], y = "Deaths", x="State/UnionTerritory", linewidth = 2, edgecolor = 'black')

plt.xlabel("Wilayah")
plt.ylabel("Total Kasus Kematian")
plt.show()
```

```python
#kasus pertumbuhan covid

fig = plt.figure(figsize=(12, 6))
ax = sns.lineplot(data=covid19_df[covid19_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])], x='Date', y='Active_cases', hue='State/UnionTerritory')
ax.set_title("Top 5 Wilayah yang Terkena Dampak di India", size=16)

plt.show()
```

```python
#Menghapus nama wilayah India

vaccine = vaccine_df[vaccine_df['State'] != 'India']
```

```python
vaccine
```

```python
vaccine.rename(columns = {"Total Individuals Vaccinated": "Total"}, inplace = True)
```

```python
#Daerah dengan jumlah vaksinasi tertinggi di India

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending = False)[:5]
max_vac 
```

```python
fig = plt.figure(figsize = (10,5))
plt.title('Top 5 Wilayah dengan Jumlah Vaksinasi Tertinggi di India', size = 20)
x = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth=2, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
